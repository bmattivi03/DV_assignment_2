# Bearing Fault Diagnosis - Feature Extraction for Clustering

**Dataset**: HUST Bearing (99 files, 51.2 kHz)  
**Goal**: Extract features and create CSV ready for clustering analysis

**Fault Types**: I (Inner), O (Outer), B (Ball), IO, IB, OB  
**Output**: CSV file with features, scaled and ready for K-means/DBSCAN

## 1. Imports

```python
import os
import glob
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import signal, stats
from scipy.io import loadmat
from scipy.fft import fft, fftfreq
from scipy.signal import butter, sosfilt, stft, welch
import pywt
from tqdm import tqdm
from sklearn.preprocessing import StandardScaler, LabelEncoder
import warnings
warnings.filterwarnings('ignore')

plt.style.use('seaborn-v0_8-darkgrid')
%matplotlib inline
np.random.seed(42)

print("âœ“ Libraries loaded")
```

## 2. Load Data

```python
# UPDATE THIS PATH
DATA_DIR = "/Users/brandomattivi/Documents/UNI/UNIBZ/MSC/1st_semester/Data_Visualization/DS_Assignment_2/DV_assignment_2/data/row_data"

def load_data(data_dir):
    mat_files = glob.glob(os.path.join(data_dir, "*.mat"))
    if len(mat_files) == 0:
        raise FileNotFoundError(f"No files in {data_dir}")
    
    records = []
    for fpath in tqdm(mat_files, desc="Loading"):
        fname = os.path.basename(fpath)
        try:
            mat = loadmat(fpath)
            sig = mat['data'].flatten()
            fs = mat.get('fs', np.array([[51200]]))[0][0]
            
            # Parse filename: I402.mat -> I=fault, 4=bearing 6204, 2=200W
            base = fname.replace('.mat', '')
            if base[1].isdigit():
                fault = base[0]
                bearing_load = base[1:]
            else:
                fault = base[:2]
                bearing_load = base[2:]
            
            bearing = f"620{bearing_load[0]}" if bearing_load[0].isdigit() else "unknown"
            load = bearing_load[1] if len(bearing_load) > 1 else "0"
            
            records.append({
                'filename': fname,
                'signal': sig,
                'fs': fs,
                'fault_type': fault,
                'bearing_type': bearing,
                'load_level': load
            })
        except Exception as e:
            print(f"Error {fname}: {e}")
    
    return pd.DataFrame(records)

df = load_data(DATA_DIR)
print(f"\nâœ“ Loaded {len(df)} signals")
print(f"  Fault types: {df['fault_type'].unique()}")
print(f"  Distribution:\n{df['fault_type'].value_counts()}")
```

## 3. Preprocess Signals

```python
def preprocess(sig):
    sig = sig - np.mean(sig)
    sig = sig / (np.std(sig) + 1e-10)
    return sig

df['signal_clean'] = [preprocess(s) for s in df['signal']]
print("âœ“ Signals preprocessed (normalized)")
```

## 4. Feature Extraction Functions

```python
def time_features(sig):
    """15 time-domain features"""
    f = {}
    f['mean'] = np.mean(sig)
    f['std'] = np.std(sig)
    f['rms'] = np.sqrt(np.mean(sig**2))
    f['peak'] = np.max(np.abs(sig))
    f['p2p'] = np.ptp(sig)
    f['abs_mean'] = np.mean(np.abs(sig))
    f['root_amp'] = np.mean(np.sqrt(np.abs(sig)))**2
    f['crest'] = f['peak'] / (f['rms'] + 1e-10)
    f['impulse'] = f['peak'] / (f['abs_mean'] + 1e-10)
    f['shape'] = f['rms'] / (f['abs_mean'] + 1e-10)
    f['clearance'] = f['peak'] / (f['root_amp'] + 1e-10)
    f['skew'] = stats.skew(sig)
    f['kurt'] = stats.kurtosis(sig)
    f['variance'] = np.var(sig)
    f['energy'] = np.sum(sig**2)
    return f

def freq_features(sig, fs):
    """10 frequency-domain features"""
    f = {}
    N = len(sig)
    fft_mag = np.abs(fft(sig))[:N//2]
    freqs = fftfreq(N, 1/fs)[:N//2]
    psd = fft_mag**2 / N
    psd_norm = psd / (np.sum(psd) + 1e-10)
    
    f['centroid'] = np.sum(freqs * psd_norm)
    f['spread'] = np.sqrt(np.sum(((freqs - f['centroid'])**2) * psd_norm))
    f['skew'] = np.sum(((freqs - f['centroid'])**3) * psd_norm) / (f['spread']**3 + 1e-10)
    f['kurt'] = np.sum(((freqs - f['centroid'])**4) * psd_norm) / (f['spread']**4 + 1e-10)
    f['dom_freq'] = freqs[np.argmax(psd)]
    f['mean_freq'] = np.sum(freqs * psd_norm)
    f['rms_freq'] = np.sqrt(np.sum((freqs**2) * psd_norm))
    
    # Energy bands
    f['low_energy'] = np.sum(psd[freqs < 1000]) / (np.sum(psd) + 1e-10)
    f['mid_energy'] = np.sum(psd[(freqs >= 1000) & (freqs < 3000)]) / (np.sum(psd) + 1e-10)
    f['high_energy'] = np.sum(psd[freqs >= 3000]) / (np.sum(psd) + 1e-10)
    
    return f

def wavelet_features(sig):
    """Wavelet energy features"""
    f = {}
    coeffs = pywt.wavedec(sig, 'db4', level=4)
    for i, c in enumerate(coeffs):
        f[f'wv{i}_energy'] = np.sum(c**2)
        f[f'wv{i}_std'] = np.std(c)
    energies = np.array([np.sum(c**2) for c in coeffs])
    f['wv_entropy'] = -np.sum((energies/np.sum(energies)) * np.log2(energies/np.sum(energies) + 1e-10))
    return f

def extract_features(sig, fs):
    """Extract all features from one signal"""
    features = {}
    
    # Time domain
    tf = time_features(sig)
    features.update({f't_{k}': v for k, v in tf.items()})
    
    # Frequency domain
    ff = freq_features(sig, fs)
    features.update({f'f_{k}': v for k, v in ff.items()})
    
    # Wavelet
    wf = wavelet_features(sig)
    features.update({f'w_{k}': v for k, v in wf.items()})
    
    return features

print("âœ“ Feature extraction functions defined")
```

## 5. Extract Features with Sliding Window

**Key decision**: Use 1-second windows with 50% overlap to increase dataset size  
- 10 second signal â†’ ~19 windows per signal  
- 99 signals â†’ **~1,881 samples** for clustering

```python
from tqdm import tqdm
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import KernelPCA
import umap

# --- Feature extraction per finestre ---
def segment_and_extract(sig, fs, window_sec=10.0, overlap=0.05):
    """Segment signal and extract features from each window"""
    window_size = int(window_sec * fs)
    step = int(window_size * (1 - overlap))
    
    windows_features = []
    for i in range(0, len(sig) - window_size + 1, step):
        window = sig[i:i + window_size]
        feats = extract_features(window, fs)
        feats['window_idx'] = i // step
        windows_features.append(feats)
    
    return windows_features


# --- Estrazione feature da tutti i segnali ---
print("ðŸ” Extracting features (this may take 2â€“3 minutes)...")
all_features = []

for idx, row in tqdm(df.iterrows(), total=len(df), desc="Processing signals"):
    windows = segment_and_extract(row['signal_clean'], row['fs'])
    
    for w in windows:
        w['filename'] = row['filename']
        w['fault_type'] = row['fault_type']
        w['bearing_type'] = row['bearing_type']
        w['load_level'] = row['load_level']
    
    all_features.extend(windows)

df_features = pd.DataFrame(all_features)

print(f"\nâœ“ Feature extraction complete!")
print(f"  Total samples: {len(df_features)}")
print(f"  Raw features per sample: {len([c for c in df_features.columns if c.startswith(('t_', 'f_', 'w_'))])}")
print(f"\n  Fault distribution:")
print(df_features['fault_type'].value_counts())


# --- Normalizzazione e riduzione dimensionale ---
print("\nâš™ï¸ Applying nonlinear dimensionality reduction...")

# Seleziona solo le colonne numeriche (feature)
feature_cols = [c for c in df_features.columns if c.startswith(('t_', 'f_', 'w_'))]
X = df_features[feature_cols].fillna(0).to_numpy(dtype=np.float64)

# Normalizzazione standard
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# --- Scegli metodo di riduzione ---
# Metodo 1: Kernel PCA (preserva strutture non lineari)
use_umap = True  # Cambia a False se vuoi usare KernelPCA

if use_umap:
    reducer = umap.UMAP(
        n_neighbors=15,
        min_dist=0.1,
        n_components=10,
        metric='euclidean',
        random_state=42
    )
    X_reduced = reducer.fit_transform(X_scaled)
    method_used = "UMAP"
else:
    kpca = KernelPCA(
        n_components=10,
        kernel='rbf',
        gamma=0.01,
        random_state=42
    )
    X_reduced = kpca.fit_transform(X_scaled)
    method_used = "KernelPCA"

# --- Creazione DataFrame ridotto ---
df_reduced = pd.DataFrame(X_reduced, columns=[f"comp_{i+1}" for i in range(X_reduced.shape[1])])
df_final = pd.concat([
    df_features[['filename', 'fault_type', 'bearing_type', 'load_level']].reset_index(drop=True),
    df_reduced
], axis=1)

# --- Salvataggio ---
df_final.to_csv("features_reduced_nonlinear.csv", index=False)

print(f"\nâœ… Nonlinear feature reduction complete!")
print(f"  Original features: {len(feature_cols)} â†’ Reduced to: {X_reduced.shape[1]}")
print(f"  Method used: {method_used}")
print(f"  Final dataset size: {df_final.shape}")
```

## 6. Data Quality Check

```python
# Check for issues
print("Data Quality Check:")
print(f"  Shape: {df_features.shape}")
print(f"  Missing values: {df_features.isna().sum().sum()}")
print(f"  Infinite values: {np.isinf(df_features.select_dtypes(include=[np.number])).sum().sum()}")

# Replace inf with NaN, then fill with median
df_features.replace([np.inf, -np.inf], np.nan, inplace=True)
numeric_cols = df_features.select_dtypes(include=[np.number]).columns
for col in numeric_cols:
    if df_features[col].isna().any():
        df_features[col].fillna(df_features[col].median(), inplace=True)

print("\nâœ“ Data cleaned")
```

## 7. Prepare for Clustering

```python
# Separate features from metadata
feature_cols = [c for c in df_features.columns if c.startswith(('t_', 'f_', 'w_'))]
meta_cols = ['filename', 'window_idx', 'fault_type', 'bearing_type', 'load_level']

X = df_features[feature_cols].values
print(f"Feature matrix: {X.shape}")

# Encode labels (for evaluation later)
le_fault = LabelEncoder()
fault_labels = le_fault.fit_transform(df_features['fault_type'])

print(f"\nFault label mapping:")
for i, name in enumerate(le_fault.classes_):
    print(f"  {name} -> {i}")

# Scale features (CRITICAL for clustering)
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

print(f"\nâœ“ Features scaled")
print(f"  Mean: {X_scaled.mean():.6f} (should be ~0)")
print(f"  Std: {X_scaled.std():.6f} (should be ~1)")
```

## 8. Visualize Key Features

```python
# Plot distribution of key features by fault type
key_feats = ['t_rms', 't_kurt', 't_crest', 'f_centroid', 'f_dom_freq', 'w_wv0_energy']

fig, axes = plt.subplots(2, 3, figsize=(15, 8))
axes = axes.flatten()

for i, feat in enumerate(key_feats):
    if feat in df_features.columns:
        df_features.boxplot(column=feat, by='fault_type', ax=axes[i])
        axes[i].set_title(feat)
        axes[i].set_xlabel('')

plt.suptitle('Feature Distributions by Fault Type', y=1.02)
plt.tight_layout()
plt.show()
```

## 9. Save Datasets

```python
# Create final dataset with scaled features + labels
df_final = pd.DataFrame(X_scaled, columns=feature_cols)
df_final['fault_type'] = df_features['fault_type'].values
df_final['fault_label'] = fault_labels
df_final['bearing_type'] = df_features['bearing_type'].values
df_final['load_level'] = df_features['load_level'].values
df_final['filename'] = df_features['filename'].values
df_final['window_idx'] = df_features['window_idx'].values

# Save CSV
output_file = "final_df.csv"
df_final.to_csv(output_file, index=False)

print(f"âœ“ Saved: {output_file}")
print(f"  Size: {os.path.getsize(output_file) / 1024:.2f} KB")
print(f"  Shape: {df_final.shape}")
print(f"\n  Columns:")
print(f"    - Features: {len(feature_cols)} (scaled, ready for clustering)")
print(f"    - Labels: fault_type, fault_label, bearing_type, load_level")
print(f"    - Metadata: filename, window_idx")

# Also save numpy array for direct use
#np.save("features_scaled_10s.npy", X_scaled)
#np.save("labels_10s.npy", fault_labels)
#print(f"\nâœ“ Also saved: features_scaled.npy, labels.npy")
```

## 10. Quick Clustering Demo (Optional)

```python
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score, adjusted_rand_score
from sklearn.decomposition import PCA

# Try K-means with 6 clusters (6 fault types)
kmeans = KMeans(n_clusters=6, random_state=42, n_init=10)
clusters = kmeans.fit_predict(X_scaled)

# Evaluate
sil_score = silhouette_score(X_scaled, clusters)
ari_score = adjusted_rand_score(fault_labels, clusters)

print(f"K-Means Clustering (k=6):")
print(f"  Silhouette Score: {sil_score:.3f} (higher is better, max=1)")
print(f"  Adjusted Rand Index: {ari_score:.3f} (agreement with true labels, max=1)")

# Visualize with PCA
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_scaled)

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14, 5))

# True labels
scatter1 = ax1.scatter(X_pca[:, 0], X_pca[:, 1], c=fault_labels, cmap='tab10', alpha=0.6, s=10)
ax1.set_title('True Fault Types (PCA projection)')
ax1.set_xlabel('PC1')
ax1.set_ylabel('PC2')
plt.colorbar(scatter1, ax=ax1)

# Predicted clusters
scatter2 = ax2.scatter(X_pca[:, 0], X_pca[:, 1], c=clusters, cmap='tab10', alpha=0.6, s=10)
ax2.set_title('K-Means Clusters (PCA projection)')
ax2.set_xlabel('PC1')
ax2.set_ylabel('PC2')
plt.colorbar(scatter2, ax=ax2)

plt.tight_layout()
plt.show()

print(f"\nâœ“ PCA explains {pca.explained_variance_ratio_.sum()*100:.1f}% variance with 2 components")
```
